```python
main = r"C:\Users\MinCheol\OneDrive\바탕 화면\STOB\STimgdataset"
Train_dir_la = r"\Train\LabeledDATA"
Test_dir_la = r"\Test\LabeledDATA"
Train_labal = main + Train_dir_la
Test_label = main + Test_dir_la


Train_path = r"C:\Users\MinCheol\OneDrive\바탕 화면\STOB\STimgdataset\Train"
dir_list = os.listdir(Train_path)

Test_path = r"C:\Users\MinCheol\OneDrive\바탕 화면\STOB\STimgdataset\Test"
di_list = os.listdir(Test_path)

print(dir_list)
print(Train_labal)

print(di_list)
print(Test_label)
```

```python
Train_paths = os.listdir(Train_labal)
print(Train_paths)

Test_paths = os.listdir(Test_label)
print(Test_paths)
```

```python
train_list = []
for TP in Train_paths:
    pa = os.path.join(Train_labal, TP)
    train_list.append(pa)
print(train_list)

test_list = []
for EP in Test_paths:
    ph = os.path.join(Test_label, EP)
    test_list.append(ph)
print(test_list)

```

```python
import xml.etree.ElementTree as ET
import pandas as pd
list_a = []
for lix in train_list:
    tree = ET.parse(lix)
    root = tree.getroot()
    for object in root.findall('object'):
        for name in object.findall('name'):
            obj_name = name.text
            if(obj_name == "othe objects"):
                obj_name = "other objects"
            if(obj_name == "other objects"):
                continue
            if(obj_name == "bundle of ropes"):
                obj_name = "bundle of rope"
            list_a.append(obj_name)
            b = set(list_a)
print(b)        #object 종류
print(len(b))   #object 개수

        
```

```python
list_b = []
obj_counts = {}
for lix in train_list:
    tree = ET.parse(lix)
    root = tree.getroot()

    for object in root.findall('object'):
        for name in object.findall('name'):
            obj_name = name.text
            if(obj_name == "othe objects"):
                obj_name = "other objects"
            if(obj_name == "other objects"):
                continue
            if(obj_name == "bundle of ropes"):
                obj_name = "bundle of rope"

            obj_counts[obj_name]= obj_counts.get(obj_name, 0) + 1
            list_b.append(obj_name)
            
for name, count in obj_counts.items():
    print(f"{name}의 모델의 Training영상의 갯수는 : {count}")
```

```python
list_c = []
obj_counts = {}
for lix in test_list:
    tree = ET.parse(lix)
    root = tree.getroot()

    for object_2 in root.findall('object'):
        for name2 in object_2.findall('name'):
            obj_names = name2.text
            if(obj_names == "othe objects" or obj_names == "other objects"):
                continue
            if(obj_names == "bundle of ropes"):
                obj_names = "bundle of rope"
            
            obj_counts[obj_names]= obj_counts.get(obj_names, 0) + 1
            list_c.append(obj_names)
            
for name, count in obj_counts.items():
    print(f"{name}의 모델의 Test영상의 갯수는 : {count}")
```
